```python
from sklearn.preprocessing import OneHotEncoder
cols = ["legal_ownership_status", "plan_configuration", 
       "position", "roof_type", "other_floor_type", 
       "foundation_type", "land_surface_condition",
       "geo_level_id_1"]

# drop building_id 

object_cols = ["land_surface_condition", 
       "foundation_type",
       "ground_floor_type",
       "other_floor_type",
       "position", 
        "legal_ownership_status",
        "roof_type", "plan_configuration"]

int64_cols = [col for col in train.columns if col not in object_cols]
```

```python
one = OneHotEncoder().fit(train[object_cols])
one_values = one.transform(train[object_cols]).toarray()
```

```python
object_values = pd.DataFrame( one_values, columns=one.get_feature_names() )
object_values["building_id"] = train["building_id"]
```

```python
train_clean = object_values.join(train[int64_cols], how="left", on="building_id",
                  lsuffix="objective")
```

```python
for x in train_clean.columns:
    print(x)
```

```python
train_clean = train_clean.drop(["building_idobjective", "building_id"], axis=1)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

train_X = train_clean

clf = GaussianNB()
clf.fit(X=train_X, y=train_Y)
```

```python
df2 = df.dropna()
```

```python
data_X = df2.drop("Y", axis=1)
data_Y = df2["Y"]

clf = GaussianNB()
clf.fit(X=data_X, y=data_Y)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, data_X, data_Y, cv=10)
print('Scores:', scores)
print('Accuracy:', scores.mean()*100)
```

```python
from sklearn.naive_bayes import MultinomialNB

data_X = df2.drop("Y", axis=1)
data_Y = df2["Y"]

clf = MultinomialNB()
clf.fit(X=data_X, y=data_Y)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, data_X, data_Y, cv=10)
print('Scores:', scores)
print('Accuracy:', scores.mean()*100)
```
